```python
# For example, rows containing missing values ​​can be deleted
df=df.dropna()
```

```python
#Empty rows of 541909 were deleted Up to 406829 rows
df                  
```

```python
# Convert date column to datetime type
df.loc[:, 'InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Convert date column to datetime type using loc
df.loc[:, 'InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
```

```python
# Filter out missing values ​​in the InvoiceDate column
df = df.dropna(subset=['InvoiceDate'])
```

```python
# Add column for year and month using loc
df.loc[:, 'YearMonth'] = df['InvoiceDate'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df.head()
```

```python
 # Calculate total sales for each month
df.groupby('YearMonth')['Quantity'].sum()
    
```

```python
monthly_sales = df.groupby('YearMonth')['Quantity'].sum()
```

```python
# Line drawing of monthly sales
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales)
plt.title('Monthly Quantity')
plt.xlabel('Month')
plt.ylabel('Quantity')
plt.show()
```

```python
# Calculate the number of customers for each month
monthly_customers = df.groupby('YearMonth')['CustomerID'].nunique()
```

```python
# Line drawing of the number of monthly customers
plt.figure(figsize=(12, 6))
plt.plot(monthly_customers)
plt.title('Monthly Customers')
plt.xlabel('Month')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Calculate the number of units sold for each product

product_sales = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales.values, y=product_sales.index)
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Product')
plt.show()
```
